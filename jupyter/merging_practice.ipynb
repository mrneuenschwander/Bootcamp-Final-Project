```python
import pandas as pd
import numpy as np
```

```python
# Read through the csv's
mvp_df = pd.read_csv('../resources/mvp_votings.csv')
nba_df = pd.read_csv('../resources/nba_info.csv')
```

```python
mvp_df.head()
```

```python
nba_df.head()
```

```python
# drop un-needd columns
clean_mvp = mvp_df.drop(columns=['ws_per_48', 'ws', 'trb_per_g', 'bpm', 'usg_pct', 'ts_pct', 'per'], axis=1)
clean_mvp.head()
```

```python
# rename columns to match both CSV's
clean_col_mvp = clean_mvp.rename(columns={'fga': 'FGM', 'fg3a': 'FG3M', 'fta': 'FTM', 'player': 'PLAYER_NAME', 'mp_per_g': 'MIN', 'pts_per_g': "PTS", 'ast_per_g': 'AST', 'stl_per_g': 'STL',
    'blk_per_g': 'BLK', 'fg_pct': 'FG_PCT', 'fg3_pct': 'FG3_PCT', 'ft_pct': 'FT_PCT'})
clean_col_mvp.head()
```

```python
nba_df.head()
```

```python
clean_nba_df = nba_df.drop(columns=['HOME_TEAM_WINS'])
clean_nba_df.head()
```

```python
# OUTER JOIN the two DataFrames together
results = pd.concat([clean_nba_df, clean_col_mvp], axis=1)
display(results)
```

```python
results.isnull().sum()
```

```python
print(results.columns.values)
```

```python
# INNER JOIN the two DataFrames together
join_results = pd.concat([clean_nba_df, clean_col_mvp], axis=1, join='inner')
display(join_results)
```

```python
print(join_results.columns.values)
```
